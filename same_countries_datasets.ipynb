```python
import pandas as pd
import numpy as np
```

```python
df_flights = pd.read_csv("converted_datasets/flights_new.csv", encoding="utf-8", encoding_errors="replace") 
print(df_flights.dtypes)
print(f"NaN Values: {df_flights.isna().sum().sum()}")
```

```python
df_restaurants = pd.read_csv("converted_datasets/restaurants_new.csv", encoding="utf-8", encoding_errors="replace") 
print(df_restaurants.dtypes)
print(f"NaN Values: {df_restaurants.isna().sum().sum()}")
```

```python
df_cities = pd.read_csv("converted_datasets/cities_new.csv", encoding="utf-8", encoding_errors="replace") 
print(df_cities.dtypes)
print(f"NaN Values: {df_cities.isna().sum().sum()}")
```

```python
df_hotels = pd.read_csv("converted_datasets/hotels_new.csv", encoding="utf-8", encoding_errors="replace") 
print(df_hotels.dtypes)
print(f"NaN Values: {df_hotels.isna().sum().sum()}")
```

```python
# Get the unique country names from df_cities
valid_countries = set(df_cities["Country"])

# Filter df_flights to keep only rows where "Country Dest" exists in df_cities["Country"]
df_flights = df_flights[df_flights["Country Arr"].isin(valid_countries)]

# Print the filtered dataframe
print(df_flights)
```

```python
print(len(sorted(df_flights["Country Arr"].dropna().unique())))
print(len(sorted(df_flights["Country Dep"].dropna().unique())))
```

```python
# Find common cities in all three dataframes
common_cities = (
    set(df_restaurants["City"])
    .intersection(set(df_cities["City"]))
    .intersection(set(df_hotels["City"]))
)

print(common_cities)
```

```python
# Find common countries in all three dataframes
common_countries = (
    set(df_restaurants["Country"])
    .intersection(set(df_cities["Country"]))
    .intersection(set(df_hotels["Country"]))
)

print(common_countries)
```

```python
# Filter all DataFrames to keep only rows where the "Country" column is in common_country
df_restaurants = df_restaurants[df_restaurants["Country"].isin(common_countries)]
df_cities = df_cities[df_cities["Country"].isin(common_countries)]
df_hotels = df_hotels[df_hotels["Country"].isin(common_countries)]
```
